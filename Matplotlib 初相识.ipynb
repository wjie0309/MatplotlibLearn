## Matplotlib 的基础

### 绘图入门

Matplotlib 是一个 Python 2D 绘图库, 能够以多种硬拷贝格式和跨平台的交互式环境生成出版物质量的图形, 用来绘制各种静态, 动态, 交互式的图表. 

现在先进行基础包的引用, 这里的简称都是平时公认度较高的缩写.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

要创建一个 Matplotlib 图像, 第一步是创建一个 figure, 每一个 figure 又包含一个或多个 axes, 这是一个可以指定坐标系的子区域.

最简单的创建 figure 以及 axes 的办法是通过 `pyplot.subplots` 命令, 在创建了之后, 可以使用 `Axes.plot` 绘制最简易的折线图.

```python
fig, ax = plt.subplots() # 创建一个包含一个 axes 的 figure
ax.plot([1,2,3,4], [1,4,2,3])
```

### Figure 的组成

Figure 是 Matplotlib 最基础的呈现方式. 通过更加完整地理解 figure 的构成, 能让我们对 Matplotlib 的功能有更全面的认识.

一个完整的 figure 通常包括四个层级, 也可以被称为容器 (container). 在使用 Matplotlib 的过程中, 我们通过各种命令方法来操控图像中的各个部分, 从而达到数据可视化的最终效果. 一副完整的图像实际上是各类子元素的集合.

	- Figure: 顶层级, 容纳所有元素
	- Axes: Matplotlib 宇宙的核心, 可以容纳大量元素并构成子图, 一个 figure 可以由一个或多个 axes 组成.
	- Axis: Axes 的下属层级, 用于处理所有和坐标轴, 网格有关的元素.
	- Tick: Axis 的下属层级, 用来处理刻度相关的元素.

### 两种绘图接口

- Matplotlib 提供两种绘图接口方式:
	- 通过前文提到的方法显示创建 figure 和 axes, 这种模式被称为 OO 模式 (object-oriented style)
	- 依赖 pyplot 自动创建 figure 和 axes.

第一种绘图接口

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple plot')
ax.legend()
plt.show()
```

第二种绘图接口

```python
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear') 
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
plt.show()
```

```python
# step1 准备数据
x = np.linspace(0, 2, 100)
y = x**2

# step2 设置绘图样式，这一模块的扩展参考第五章进一步学习，这一步不是必须的，样式也可以在绘制图像是进行设置
mpl.rc('lines', linewidth=4, linestyle='-.')

# step3 定义布局， 这一模块的扩展参考第三章进一步学习
fig, ax = plt.subplots()  

# step4 绘制图像， 这一模块的扩展参考第二章进一步学习
ax.plot(x, y, label='linear')  

# step5 添加标签，文字和图例，这一模块的扩展参考第四章进一步学习
ax.set_xlabel('x label') 
ax.set_ylabel('y label') 
ax.set_title("Simple Plot")  
ax.legend()
```
