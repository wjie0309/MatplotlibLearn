## Artist

### 概述

#### Matplotlib 的三层 API

Matplotlib 的实现原理与人在画布上画画相似, 使用 Artist 在画布 (canvas) 上绘制 (Render) 图形.

基于这个原理, matplotlib 有三个层次的 API:

	- matplotlib.backend_bases.FigureCanvas: 画布 (绘图区)
	- matplotlib.backend_bases.Renderer: 画笔 (渲染器, 控制如何在 canvas 上画图)
	- matplotlib.artist.Artist 代表了具体的图表组件, 即调用画笔在画布上画画的方法

我们在使用 matplotlib 的过程中, 大部分时间都是使用 Artist 类的 api.

#### Artist 的分类

Artist 被分为两种类型: `primitive` 和 `containers`.

`primitive` 是基本要素, 其包含我们在作图过程中的基本对象, 如线条, 图像, 矩形等.

`containers` 是承载基本要素的地方, 如 figure, axes, axis 等.

### Primitive

现重要介绍 `primitive` 的几个重要类型, 曲线-line2D, 矩形-rectangle, 多边形-polygon, 图像-image

#### Line2D

在绘制曲线的时候, 主要使用 `matplotlib.lines.Line2D` 实现.

- 构造函数:

	```
	class matplotlib.lines.Line2D(xdata, ydata, linewidth=None, linestyle=None, color=None, marker=None, markersize=None, markeredgewidth=None, markeredgecolor=None, markerfacecolor=None, markerfacecoloralt='none', fillstyle=None, antialiased=None, dash_capstyle=None, solid_capstyle=None, dash_joinstyle=None, solid_joinstyle=None, pickradius=5, drawstyle=None, markevery=None, **kwargs)
	```

- 其中常用的参数:

	- xdata: 点在 x 轴上的取值
	- ydata: 点在 y 轴上的取值
	- linewidth
	- linestyle
	- color
	- marker: 点的标记类型
	- markersize

- 如何设置 Line2D 的属性

	有三种方法设置曲线的属性

	- 直接在 plot() 函数时设置
	- 通过获得线对象, 然后对其进行设置
	- 使用 step() 函数对线对象进行设置

````python
# 直接设置
x = range(0, 5)
y = [2, 5, 7, 8, 10]
plt.plot(x, y, linewidth=10)
````

````python
# 获得线对象, 对其进行设置
x = range(0,5)
y = [2, 5, 7, 8, 10]
line, = plt.plot(x, y, '-') # line, 是一个列表的解包操作, 从而获得line2D返回的对象
line.set_antialiased(False)
````

````python
# 使用 set() 来设置线对象
x = range(0,5)
y = [2, 5, 7, 8, 10]
line = plt.plot(x, y, '-') # line, 是一个列表的解包操作, 从而获得line2D返回的对象
plt.setp(line, linewidth=10, color='r')
````

#### 如何绘制 lines

1. 绘制直线 line
2. errorbar 绘制误差折线图

##### 绘制 line 常用的方法同样有两种:
- plot 方法绘制
- Line2D 对象绘制

````python
# plot方法绘制

x = range(0,5)
y1 = [2,5,7,8,10]
y2= [3,6,8,9,11]

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
print(ax.lines)  # 打印lines信息可以看见两个Line2D对象被创建了
````

````python
# Line2D对象绘制

x = range(0,5)
y1 = [2,5,7,8,10]
y2= [3,6,8,9,11]

fig, ax = plt.subplots()
lines = [Line2D(x, y1), Line2D(x, y2, color='orange')]  # 显式创建对象
# 添加到 ax 中
for line in lines:
    ax.add_line(line)
ax.set_xlim(0,4)
ax.set_ylim(2,11)
print(ax.lines)
````

##### errorbar 绘制误差折线图

pyplot里有个专门绘制误差线的功能，通过`errorbar`类实现，它的构造函数：

> matplotlib.pyplot.errorbar(x, y, yerr=None, xerr=None, fmt='', ecolor=None, elinewidth=None, capsize=None, barsabove=False, lolims=False, uplims=False, xlolims=False, xuplims=False, errorevery=1, capthick=None, *, data=None, **kwargs)

- 其中最主要的几个参数为:

	- x
	- y
	- xerr: 指定 x 轴水平的误差
	- yerr: 指定 y 轴水平的误差
	- fmt: 指定折线图中某个点的颜色, 形状, 线条风格等, e.g. 'co--'
	- ecolor: error bar 的颜色
	- elinewidth: error bar 的线条宽度

````python
fig = plt.figure()
x = np.arange(10)
y = 2.5 * np.sin(x / 20 * np.pi)
yerr = np.linspace(0.05, 0.2, 10)
plt.errorbar(x, y, yerr=yerr, fmt='o-', ecolor='r', elinewidth=2)
````

#### Patches

`matplotlib.patches.Patch` 类是二维图形类, 是众多二维图形的父类.

本文着重介绍常用的矩形, 多边形和楔形.

##### 矩形 Rectangle

构造函数如下:

> class matplotlib.patches.Rectangle(xy, width, height, angle=0.0, **kwargs)

对于矩形类, 通过锚点 xy 和宽高来确定矩形. 最常用的矩形类是 hist 和 bar.

###### Hist

> matplotlib.pyplot.hist(x,bins=None,range=None, density=None, bottom=None, histtype='bar', align='mid', log=False, color=None, label=None, stacked=False, normed=None)

- 常用参数:
	- x
	- bins: 统计的区间分布
	- range: tuple, 显示的区间
	- density: 显示频率统计结果
	- histtype: {'bar', 'barstacked', 'step', 'stepfilled'} 之一, 为矩形的不同填充方式
	- align: 对齐方式
	- log: y 轴是否使用指数刻度
	- stacked: 是否为堆积状图

````python
# hist 绘制直方图

x = np.random.randint(0, 100, 100)  # 0-100的100个随机数
bins = np.arange(0, 101, 10) # 生成边界值, 这里同样注意边界值是到100, 所以 arange 取到 101
plt.hist(x, bins, color='fuchsia', alpha=0.3)  # alpha 调节透明度
plt.xlabel('scores')
plt.ylabel('count')
# plt.xlim(0, 100)  # 设置分布范围
plt.show()
````

###### Bar

> matplotlib.pyplot.bar(left, height, alpha=1, width=0.8, color=, edgecolor=, label=, lw=3)

- 常用参数

	- height: y 轴的数值序列
	- alpha: 透明度
	- width: 宽度
	- color
	- edgecolor: 图形边缘颜色
	- label: 每个图像代表的含义

##### 多边形 Polygon

`matplotlib.patches.Polygon` 类是多边形类, 构造函数为


> class matplotlib.patches.Polygon(xy, closed=True, **kwargs)


xy是一个 N×2 的 numpy array, 为多边形的顶点.
closed 为是否指定多边形起点与终点重合而封闭多边形.

在 polygon 类中, 最常用的是 fill 类, 用于绘制一个填充的多边形.

> matplotlib.pyplot.fill(*args, data=None, **kwargs)


参数: x, y, color

````python
x = np.linspace(0, 5*np.pi, 1000)
y1 = np.sin(x)
y2 = np.sin(2*x)

# 可以通过多次 fill 来添加多个多边形
plt.fill(x, y2, color='g', alpha=0.3)
plt.fill(x, y1, color='grey', alpha=0.3)
````

#### 楔形 Wedge

Wedge 类的构造函数:

> class matplotlib.patches.Wedge(center, r, theta1, theta2, width=None, **kwargs)

可见 wedge 的构造方法是给定 center, 半径 r, 以及扫过的起始角度和结束角度. wedge 中比较常见的是绘制饼状图.

`matplotlib.pyplot.pie` 的构造函数:

> matplotlib.pyplot.pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=0, radius=1, counterclock=True, wedgeprops=None, textprops=None, center=0, 0, frame=False, rotatelabels=False, *, normalize=None, data=None)

- 主要参数
	- x: 楔形的形状, 一维数组
	- explode: 为每一个楔形块偏离的程度
	- labels: 每个楔形快的标记.
	- colors: 饼图循环使用的颜色
	- starangle: 第一个楔形块开始的角度

````python
# pie绘制饼状图

labels = ['1', '2', '3', '4']
sizes = [20, 35, 15, 30]
explode = (0, 0.15, 0, 0.1)

plt.pie(sizes, explode, labels,autopct='%1.1f%%', shadow=True, startangle=90)
````

#### Images

Images 是 matplotlib 中绘制 image 的类, 其中最常见的为 imshow 类, 可以根据数组绘制为图像.

> class matplotlib.image.AxesImage(ax, cmap=None, norm=None, interpolation=None, origin=None, extent=None, filternorm=True, filterrad=4.0, resample=False, **kwargs)

> matplotlib.pyplot.imshow(X, cmap=None, norm=None, aspect=None, interpolation=None, alpha=None, vmin=None, vmax=None, origin=None, extent=None, shape=, filternorm=1, filterrad=4.0, imlim=, resample=None, url=None, *, data=None, **kwargs）

使用imshow画图时首先需要传入一个数组, 数组中储存着空间的像素位置和像素点的位置. interpolation 可以设置为不同的差值方法.

````python
methods = [None, 'none', 'nearest', 'bilinear', 'bicubic', 'spline16',
           'spline36', 'hanning', 'hamming', 'hermite', 'kaiser', 'quadric',
           'catrom', 'gaussian', 'bessel', 'mitchell', 'sinc', 'lanczos']

grid = np.random.rand(4,4)

fig, axs = plt.subplots(nrows=3, ncols=6, figsize=(9, 6))

for ax, interp_method in zip(axs.flat, methods):
    ax.imshow(grid, interpolation=interp_method)
    ax.set_title(str(interp_method))
    
plt.tight_layout()  # 自动调整大小
````

#### Figure 容器

Figure 是最顶层的 container, 它包含了图表的所有元素.

当我们向图表添加 `Figure.add_subplot()` 或者 `Figure.add_axes()` 元素时, 这些也会被添加到 `Figure.axes` 列表中.

由于 `Figure` 维持了 `current axes` , 因此你不应该手动的从 `Figure.axes` 列表中添加删除元素, 而是要通过 `Figure.add_subplot()` , `Figure.add_axes()` 来添加元素, 通过 `Figure.delaxes()` 来删除元素.

- Figure 的常见属性
	- Figure.patch: 背景矩形
	- Figure.axes: Axes 实例的列表
	- Figure.imges
	- Figure.lines: Line2D 的实例列表
	- Figure.legends: Figure Legend 实例列表
	- Figure.text

````python
fig = plt.figure()
ax1 = fig.add_subplot(211)  # 做一副2*1的图, 选择第1个子图
ax2 = fig.add_axes([0,0.3, 0.7,0.1])  # (left,bottom,width,height)
print(fig.axes)

# 对axes添加网格

for ax in fig.axes:
    ax.grid(True)
````

#### Axes

`matplotlib.axes.Axes` 是matplotlib的核心. 大量的用于绘图的 `Artist` 存放在它内部, 并且它有许多辅助方法来创建和添加 `Artist` 给它自己, 而且它也有许多赋值方法来访问和修改这些 `Artist`.

Subplot 就是一个特殊的 Axes, 它是位于网格中的某个区域的 subplot 实现的, 我们也可以在任意区域创建 Axes, 如同上文中在 Figure 中创建 Axes 一样, 通过`Figure.add_axes([left,bottom,width,height])`来创建一个任意区域的Axes, 其中参数都是 0-1 之间的浮点数, 代表了相对于 Figure 的坐标.

在 Axes 中创建新的 line 或者 patch 时, 我们通常使用辅助方法 `axes.add_line()` 等来实现, 因为在创建新的实例时, Axes 会进行一系列的自动化工作, 直接添加的话会忽略这些步骤.

- Axes 的常见属性:

	`artists`: Artist实例列表  
	`patch`: Axes所在的矩形实例  
	`collections`: Collection实例  
	`images`: Axes图像  
	`legends`: Legend 实例  
	`lines`: Line2D 实例  
	`patches`: Patch 实例  
	`texts`: Text 实例  
	`xaxis`: matplotlib.axis.XAxis 实例  
	`yaxis`: matplotlib.axis.YAxis 实例

另外Axes还包含两个最重要的Artist container：

`ax.xaxis`: XAxis对象的实例，用于处理x轴tick以及label的绘制  
`ax.yaxis`: YAxis对象的实例，用于处理y轴tick以及label的绘制

#### Axis

`matplotlib.axis.Axis` 实例处理 `tick line`, `grid line`, `tick label` 以及 `axis label` 的绘制.

每个 Axis 都有一个 label 属性, 也有主刻度列表和次刻度列表.

刻度是动态创建的, 只有在需要创建的时候才创建(比如缩放的时候). Axis也提供了一些辅助方法来获取刻度文本, 刻度线位置等等.

````python
# 不用print，直接显示结果
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

fig, ax = plt.subplots()
x = range(0,5)
y = [2,5,7,8,10]
plt.plot(x, y, '-')

axis = ax.xaxis # axis为X轴对象
axis.get_ticklocs()     # 获取刻度线位置
axis.get_ticklabels()   # 获取刻度label列表(一个Text实例的列表）。 可以通过minor=True|False关键字参数控制输出minor还是major的tick label。
axis.get_ticklines()    # 获取刻度线列表(一个Line2D实例的列表）。 可以通过minor=True|False关键字参数控制输出minor还是major的tick line。
axis.get_data_interval()# 获取轴刻度间隔
axis.get_view_interval()# 获取轴视角（位置）的间隔
````

#### Tick

`Tick` 包含了 `tick`, `grid line` 实例以及对应的 `label`.

- Tick 的常见属性

	`Tick.tick1line`: Line2D实例  
	`Tick.tick2line`: Line2D实例

````python
import pandas as pd 

data = pd.read_csv("Drugs.csv")

state = set(data['State'])  # 找出State的类别
year = set(data['YYYY'])
year = list(year)
year.sort()
print(state, year)

print(data)

test = data.groupby(['YYYY'])['DrugReports'].sum()
print(test)
````

````python
figure, axes = plt.subplots(figsize=[9,8])
axes.grid(True, color='white')
axes.patch.set_facecolor('lightgrey')
plt.xlabel('year')
plt.ylabel('DrugReports') 


for i in state:
    tmp = data.where(data['State'] == i)
    subdata = tmp.groupby(['YYYY'])['DrugReports'].sum()
    if i == 'PA':
        plt.plot(year, subdata, linewidth=5, color='orange', label = i)
    else:
        plt.plot(year, subdata, label = i, color='grey')
        
print(figure)
````
