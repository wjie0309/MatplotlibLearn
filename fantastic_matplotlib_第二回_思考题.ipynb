# 拿数据的思路：
1.将数据按照State进行分组，可以得到每个State每年的所有数据
2.按年进行求和
3.因为YYYY是索引，直接拿数据拿不到，所以通过df['year']=df.index创造一个新列，就能通过df['year']去取数据

```python
import pandas as pd
import matplotlib
path = 'Drugs.csv'
data = pd.read_csv(path)
```
